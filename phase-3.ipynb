```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1 , 150000 , 180000 , 220000 , 250000],
    [2 , 120000 , 140000 , 160000 , 190000],
    [3 , 200000 , 230000 , 260000 , 300000],
    [4 , 180000 , 210000 , 240000 , 270000],
    [5 , 160000 , 185000 , 205000 , 230000]
])

print("======== ZOMATO SALES ANLYSIS ==========")
print("\n SALES DATA SHAPE : " , sales_data.shape)
print("\n SAMPLE DATA FOR 1st 3 ROWS \n" , sales_data[0:3])
print("\n SAMPLE DATA FOR 1st 3 ROWS \n" , sales_data[: , 1:])
```

```python
print(np.sum(sales_data[: , 1:] , axis=0))

yearly = np.sum(sales_data[: , 1:] , axis=0)
print("\n\n  ===== YEARLY TOTAL DATA ===== \n", yearly)

min_res = np.min(sales_data[: , 1:], axis=1 )
print("\n\n ===== MIIMUM SALES PER RESTURANT ===== \n ", min_res)

max_res = np.max(sales_data[:, 1:] , axis=0)
print("\n\n ===== MAXIMUM SALES PER YEAR ===== \n ", max_res)

avg_year = np.mean(sales_data[: , 1:] ,axis=1)
print("\n\n ===== AVERAGE SALES PER YEAR ===== \n ", avg_year)
```

```python
cumalative = np.cumsum(sales_data[: , 1:] , axis=1)
print("\n\n ===== CUMALATIVE SALES PER YEAR ===== \n ", cumalative)

plt.figure(figsize=(8, 5))
plt.plot(np.average(cumalative , axis = 0))
plt.title("Average Cumlative sales across all resturant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show
```
